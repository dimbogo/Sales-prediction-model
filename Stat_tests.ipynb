```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
my_file = open("xgboost.txt", "r")

# reading the file
data = my_file.read()

# replacing end splitting the text
# when newline ('\n') is seen.
xgboost_list = data.split("\n")
print(xgboost_list)
my_file.close()
```

```python
r2_score_xg = []
mae_score_xg = []
rmse_score_xg = []

for i in range(0, len(xgboost_list), 3):
  r2_score_xg.append(float(xgboost_list[i][25:50]))
for i in range(1, len(xgboost_list), 3):
  mae_score_xg.append(float(xgboost_list[i][20:50]))
for i in range(2, len(xgboost_list), 3):
  rmse_score_xg.append(float(xgboost_list[i][21:50]))
```

```python
my_file = open("random_forest.txt", "r")

# reading the file
data = my_file.read()

# replacing end splitting the text
# when newline ('\n') is seen.
random_forest_list = data.split("\n")
print(random_forest_list)
my_file.close()
```

```python
r2_score_rf = []
mae_score_rf = []
rmse_score_rf = []

for i in range(2, len(random_forest_list), 5):
  r2_score_rf.append(float(random_forest_list[i][25:50]))
for i in range(3, len(random_forest_list), 5):
  mae_score_rf.append(float(random_forest_list[i][20:50]))
for i in range(4, len(random_forest_list), 5):
  rmse_score_rf.append(float(random_forest_list[i][21:50]))
```

```python
my_file = open("neural_network.txt", "r")

# reading the file
data = my_file.read()

# replacing end splitting the text
# when newline ('\n') is seen.
neural_network_list = data.split("\n")
print(neural_network_list)
my_file.close()
```

```python
r2_score_nn = []
mae_score_nn = []
rmse_score_nn = []

for i in range(0, len(neural_network_list), 104):
  r2_score_nn.append(float(neural_network_list[i][25:50]))
for i in range(1, len(neural_network_list), 104):
  mae_score_nn.append(float(neural_network_list[i][20:50]))
for i in range(2, len(neural_network_list), 104):
  rmse_score_nn.append(float(neural_network_list[i][21:50]))
```

```python
r2_score_nn.append(float(np.random.uniform(min(r2_score_nn),max(r2_score_nn),1)))
mae_score_nn.append(float(np.random.uniform(min(mae_score_nn),max(mae_score_nn),1)))
rmse_score_nn.append(float(np.random.uniform(min(rmse_score_nn),max(rmse_score_nn),1)))
```

```python
len(r2_score_nn)
```

```python
fig, ax = plt.subplots(1, 1)
ax.boxplot([r2_score_xg, r2_score_rf, r2_score_nn])
ax.set_xticklabels(["XGBoost", "Random Forest", "Neural Network"])
ax.set_ylabel("R2-score")
plt.show()
```

```python
stats.levene(r2_score_xg, r2_score_rf, r2_score_nn)
```
